Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Data Cleansing Phase 1 
prop_transp = properties.transpose()

new_prop = prop_transp[1:]

new_prop.columns = prop_transp.iloc[0]

new_prop.reset_index(inplace = True)

print(new_prop.columns)

new_prop.head(10)
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Data Cleansing Phase 2

new_prop.columns = new_prop.columns.fillna('placeholder')

new_prop = new_prop.rename(columns = {'index': "London_Borough", 'placeholder': "Parent_Area"})

new_prop.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt the Dataframe 
clean_prop = pd.melt(new_prop, id_vars = ['London_Borough', 'Parent_Area'])
clean_prop.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Validating data types 
clean_prop.dtypes
```

```python
clean_prop['value'] = pd.to_numeric(arg = clean_prop['value'])
clean_prop.sample(5)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Data Cleansing Phase 3
unik_districts = list(clean_prop["London_Borough"].unique())
unik_districts
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
def remove_rows(df, district):
    df1 = df[df['London_Borough']==district]
    
    #checking if there is at least one null value
    if df1["value"].isnull().sum() > 0:
        #if there is, then remove all rows with that district name
        df2 = df[df['London_Borough']!=district]
        return df2
    else:
        #otherwise, leave dataframe unchanged
        return df

#Copy the dataframe
new_prop = clean_prop.copy()

for dis in unq_districts:
    new_prop = remove_rows(new_prop, dis)

list(new_prop["London_Borough"].unique())
```

# 3. Data Viz

```python
# Standardizing and renaming data columns 
new_prop = new_prop.rename(columns = {'Unnamed: 0': "Month", 'value':'Average_price'})
new_prop.head(30)
```

```python
Newham_district = new_prop[new_prop["London_Borough"] == "Newham"]
```

```python
ax = Newham_district.plot(kind="line", x="Month", y="Average_price")
ax.set_ylabel("Price")
```

```python
Hackney_district = new_prop[new_prop["London_Borough"] == "Hackney"]
ax = Hackney_district.plot(kind="line", x="Month", y="Average_price")
ax.set_ylabel("Price")
```

```python

Kensington_Chelsea_district = new_prop[new_prop["London_Borough"] == "Kensington & Chelsea"]
ax = Kensington_Chelsea_district.plot(kind="line", x="Month", y="Average_price")
ax.set_ylabel("Price")
```

```python
#Lambda function to extract year from the Month column
new_prop["Year"] = new_prop["Month"].apply(lambda y:y.year)
new_prop.head()
```

```python
new_prop_means = new_prop.groupby(by=['London_Borough', 'Year'], as_index=False).mean()
new_prop_means
```

```python
new_prop_means.reset_index()
new_prop_means.columns
```

```python
#Data Viz to understand pricing in 1998
plt.rcParams['figure.figsize'] = [15, 8]

data_1998 = new_prop_means[new_prop_means['Year']==1998]
data_1998 = data_1998.sort_values(by="Average_price", ascending=False)

plot_1998 = data_1998[['London_Borough','Average_price']].plot(kind='bar', y = "Average_price", title = "Average Prices for the year 1998")
plot_1998.set_xticklabels(data_1998.London_Borough);
```

```python
#Data Viz to understand pricing 2018
plt.rcParams['figure.figsize'] = [15, 8]

data_2018 = new_prop_means[new_prop_means['Year']==2018]
data_2018 = data_2018.sort_values(by="Average_price", ascending=False)

plot_2018 = data_2018[['London_Borough','Average_price']].plot(kind='bar', y = "Average_price", title = "Average Prices for the year 2018")
plot_2018.set_xticklabels(data_2018.London_Borough);
```

# 4. Modeling

```python
#Creating function to compare house prices in 1998 to 2018
def create_price_ratio(df):
    year_1998 = float(df['Average_price'][df["Year"] == 1998])
    year_2018 = float(df['Average_price'][df["Year"] == 2018])
    
    year_ratio = [year_2018/year_1998]
    return year_ratio
```

```python
# Creating a dictionary for the ratio where the key is district/borough 
ratio_dict = {}



for district in new_prop_means["London_Borough"].unique():
    district_data = new_prop_means[new_prop_means["London_Borough"] == district]
    
    dis_ratio = create_price_ratio(district_data)
    
    ratio_dict[district] = dis_ratio

print(ratio_dict)
```

```python
#Creating a Dataframe for the dictionary containing the ratios
mean_ratios = pd.DataFrame(ratio_dict)

mean_ratios = mean_ratios.transpose()

mean_ratios.reset_index(inplace=True)
```

```python
#column renaming
mean_ratios.rename(columns={'index':'London_Boroughs', 0:'2018_ratios'}, inplace = True)
```

```python
#Ascending order sorting for ratios 
ordered_ratios = mean_ratios.sort_values(by='2018_ratios',ascending=False)
```

```python
#Data Viz for the 2018/1998 average price ratio 

ax = ordered_ratios[['London_Boroughs','2018_ratios']].plot(kind='bar', title = "2018/1998 average price ratio of London Boroughs")
ax.set_xticklabels(ordered_ratios.London_Boroughs);
```
